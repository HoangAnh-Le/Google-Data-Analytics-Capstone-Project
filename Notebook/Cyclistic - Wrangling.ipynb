```python
import pandas as pd

Jan2025 = pd.read_csv('202501-divvy-tripdata.csv')
Feb2025 = pd.read_csv('202502-divvy-tripdata.csv')
Mar2025 = pd.read_csv('202503-divvy-tripdata.csv')
Apr2025 = pd.read_csv('202504-divvy-tripdata.csv')
May2025 = pd.read_csv('202505-divvy-tripdata.csv')
Jun2025 = pd.read_csv('202506-divvy-tripdata.csv')
```

```python
#Add all datasets to a list
dfs = [Jan2025, Feb2025, Mar2025, Apr2025, May2025, Jun2025]
```

```python
#convert ride_length to timedelta

for df in dfs:
    df['ride_length'] = pd.to_timedelta(df['ride_length'])
```

```python
#add a column 'Month' to display the month of the records
Jan2025['Month'] = 'Jan'
Feb2025['Month'] = 'Feb'
Mar2025['Month'] = 'Mar'
Apr2025['Month'] = 'Apr'
May2025['Month'] = 'May'
Jun2025['Month'] = 'Jun'
```

```python
#combine all datasets into 1
combined_df = pd.concat(dfs)
```

```python
#drop unnecessary columns
combined_df = combined_df.drop(columns=['start_lng','start_lat','end_lat','end_lng'])
```

```python
combined_df
```

```python
# Descriptive analysis on ride_length by month (all figures in seconds)

combined_df.groupby('Month').agg({
    'ride_length': ['mean', 'max', 'min', 'median']
})
```

```python
# Descriptive analysis on ride_length

combined_df.agg({
    'ride_length': ['mean', 'median', 'min', 'max']
})
```

```python
# Compare members and casual users

combined_df.groupby('member_casual').agg({
    'ride_length': ['mean', 'max', 'min', 'median']})
```

```python
#convert day_of_week to real date (Mon, tue...)

def real_date(val):
    if val == 1:
        return 'Sunday'
    elif val == 2:
        return 'Monday'
    elif val == 3:
        return 'Tuesday'
    elif val == 4:
        return 'Wednesday'
    elif val == 5: 
        return 'Thursday' 
    elif val == 6:
        return 'Friday'
    elif val == 7:
        return 'Saturday'

combined_df['day_of_week'] = combined_df['day_of_week'].apply(real_date)
```

```python
# See the average ride time by each day for members vs casual users

##Define the correct weekday order
weekday_order = [
    'Monday', 'Tuesday', 'Wednesday', 
    'Thursday', 'Friday', 'Saturday', 'Sunday'
]

## Convert the column into an ordered categorical
combined_df['day_of_week'] = pd.Categorical(
    combined_df['day_of_week'],
    categories=weekday_order,
    ordered=True
)

##pivot table 

pivot = pd.pivot_table(
    combined_df,
    values='ride_length',
    index='day_of_week',
    columns='member_casual',
    aggfunc=['mean','count'],
    fill_value=0,
    margins=True
)

# Remove All row and column
pivot = pivot.drop(index='All')
pivot = pivot.drop(columns='All', level=1)

# Flatten column names
pivot.columns = [f"{agg}_{user}" for agg, user in pivot.columns]

pivot
```

```python
#convert ride_length to minutes

combined_df['ride_length_minutes'] = combined_df['ride_length'].dt.total_seconds() / 60

combined_df['ride_length_minutes'] = combined_df['ride_length_minutes'].round(3)

combined_df = combined_df.drop(columns='ride_length')
```

```python
combined_df.to_csv("combined_df.csv", index=False)
```
